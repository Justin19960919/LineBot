```python
import json
import requests
from bs4 import BeautifulSoup
from lxml import etree
```

```python
## writing my own scraper
# https://dev.to/iankerins/the-easy-way-to-build-an-instagram-spider-using-python-scrapy-graphql-4gko

## Everday a meme
#https://www.instagram.com/everyday.a.meme/

url = 'https://www.instagram.com/everyday.a.meme/'
print("Executing....")

def get_url(url):
    try:
        response = requests.get(url)
        if response.status_code == 200:
            print("Access approved...")
            return response.text
        else:    
            print(response.status_code)        
    except Exception as e:
        print(e)
        return None

html = get_url(url)
```

```python
# getting posts
selector = etree.HTML(html)
links = selector.xpath("//script[starts-with(.,'window._sharedData')]/text()")[0]

json_string = links.strip().split("= ")[1][:-1]
data = json.loads(json_string)
data
```

```python
edges = data['entry_data']['ProfilePage'][0]['graphql']['user']['edge_owner_to_timeline_media']['edges']
url = []
for edge in edges:
    display_url = edge['node']['display_url']
    print(display_url)
    url.append(display_url)
```
